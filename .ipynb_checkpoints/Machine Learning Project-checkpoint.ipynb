## ISOLATION FOREST MODEL

```python
#ANOMALY DETECTION TROUGH ISOLATION FOREST
"""Outlier detection using the Isolation Forest algorithm applied to numeric variables. 
A 1% contamination threshold (contamination=0.01) is set to identify anomalous data. Anomalies are counted and displayed, 
but the labeling column (Anomaly) is ultimately removed to keep the dataset clean for the subsequent mode."""

#Data preparation for anomaly detection
# We use only the numeric columns to find outliers
features_for_anomaly = ['CreditScore', 'Age', 'Balance', 'Salary', 'NumOfProducts']
X_anomaly = df[features_for_anomaly] #Is my new dataframe for Isolation Forest

#Application of Isolation Forest 

iso = IsolationForest(contamination=0.01, random_state=42) #I put the contamination to 0.01, it means that we expect about 1% outliers
y_anomaly = iso.fit_predict(X_anomaly)

#To create a column in the dataframe: -1 is anomaly, 1 is normal
df['Anomaly'] = y_anomaly
anomalies = df[df['Anomaly'] == -1]

print("----- ANOMALY DETECTION (Isolation Forest) -----")
print(f"Found {len(anomalies)} anomalies on {len(df)} clients.")
print("Example of anomalous customers detected:")
print(anomalies.head()) #To print the first five anomalies

#To remove the anomaly column for the model 
df = df.drop('Anomaly', axis=1) 
print("Column 'Anomaly' removed from training dataset.")
```

We applied the Isolation Forest algorithm to detect potential anomalies in the numerical data, setting a contamination level of 1%. The algorithm identified **80** **outliers**.

A **qualitative analysis** of these cases (shown in the output) reveals that they are not data entry errors, but rather **rare** but valid user profiles: for example, customers with the maximum number of products (4) or a perfect credit score (850). Therefore, we **decided** to keep them in the dataset so that the model can also learn these extreme behaviors.

#

## DATA PREPARATION

```python
#DATA PREPARATION
"""Data Preparation phase: encoding categorical variables (Label Encoding for 'Gender'), transform the variable 'Geography' into a dummies, 
isolating the target (Churn) and the sensitive attribute (Gender). Finally, data splitting into Training Set (80%) and Test Set (20%)."""


print('----- DATA PREPARATION -----')
le = LabelEncoder() #To create the object to transform Gender in 0,1
df['Gender_Encoded'] = le.fit_transform(df['Gender'])#I use fit, to learn all the column of gender, and to understand all the variable (in this case Male,Female),and after he transform in alphabetical order the variable. 


#To transform the variable "Geography" in a dummies, and we drop the first variable to don't have redundancy
df = pd.get_dummies(df, columns=['Geography'], drop_first=True)

#Target (Variable to predict)
#Sensitive Attribute (for Fairness analysis), since for the model we have Gender_Enconded
y = df['Churn']          
S = df['Gender']         

#Features matrix (X)
#Remove Target and the original string Gender column
X = df.drop(['Churn', 'Gender'], axis=1) 


#Split the sample for the Train (80%) and the Test (20%).I use random_state=42 for reproducibility. I put S for the fairness
X_train, X_test, y_train, y_test, S_train, S_test = train_test_split(
    X, y, S, test_size=0.2, random_state=42
)



#To print the dimension of Traint set and Test set
print(f"Train Set Dimensions: {X_train.shape}")
print(f"Test Set Dimensions : {X_test.shape}")
```

#

## IMPLEMENTATION OF MODELS

#### Gradient Boosting

```python
#GRADIENT BOOSTING OPTIMIZATION
"""Gradient Boosting parameter optimization. The code iterates through various combinations
of tree counts (n_estimators), learning rates (lerning_rates) and depths (max_depth), evaluating each configuration 
with Cross-Validation (ROC-AUC metric). Finally, the model is retrained with the best configuration found."""

print("----- GRADIENT BOOSTING OPTIMIZATION  -----")

#Define the base model and fix random_state for reproducibility
gb_base = GradientBoostingClassifier(random_state=42)

#Define the parameters that i want to try
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 300],
    'max_depth': [2, 3, 4]
}

#Implementation of GridSearch
grid_search = GridSearchCV(
    estimator=gb_base,
    param_grid=param_grid,
    cv=5,                 #Evaluate with Cross-Validation (5-Fold)
    scoring='roc_auc',    #I use ROC as metrics of valutation
    n_jobs=-1,            #To velocize the process
    verbose=1,            #To give the information during the process
    refit=True            #Retrain the best at the end
)

#To fit the best model
grid_search.fit(X_train, y_train)

best_index = grid_search.best_index_                              #To take the standard deviation
best_std = grid_search.cv_results_['std_test_score'][best_index]  #To take the standard deviation of the Cross-Validation (5-Fold)

#To print the Result
print("----------------------------------------")
print(f"Best GB AUC = {grid_search.best_score_:.4f} (+/- {best_std * 2:.4f})")
print("Best GB Params:", grid_search.best_params_)


#To use the best model and i assign to gb
gb = grid_search.best_estimator_

```

**Implementation** of Gradient Boosting, testing more parameters, to optimize the parameters for Gradient Boosting. The **standard deviation (std)** remains consistently low (+/- 0.02) across nearly all tests. This indicates that the model is **highly stable**: its performance does not vary drastically depending on how the data is split between training and validation.

#

#### Random Forest

```python
#SEARCH FOR THE BEST PARAMETER FOR RANDOM FOREST
"""Random Forest parameter optimization. The code iterates through various combinations
of tree counts (n_estimators) ,depths (max_depth) and minimum samples for splitting (min_samples_leaf), evaluating each configuration with
Cross-Validation (ROC-AUC metric). The class_weight='balanced' parameter is used to handle class imbalance. Finally, the model is retrained with 
the best configuration found."""

print("----- RANDOM FOREST OPTIMIZATION  -----")

#Define the base model and fix random_state for reproducibility
rf_base = RandomForestClassifier(
    random_state=42,
    oob_score=True,           #To Calculate the Out-of-Bag score
    class_weight='balanced'   #i use Balanced to force the model to give more importance to the people that will leave

)

#Define the parameters that i want to try
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20, 25],
    'min_samples_leaf': [2, 5, 10, 20]
}

#Implementation of GridSearch
grid_search = GridSearchCV(
    estimator=rf_base,
    param_grid=param_grid,
    cv=5,                           #Evaluate with Cross-Validation (5-Fold)
    scoring='roc_auc',              #I use ROC as metrics of valutation
    n_jobs=-1,                      #To velocize the process
    verbose=1,                      #To give the information during the process
    refit=True                      #Retrain the best at the end
)

#To fit the best model
grid_search.fit(X_train, y_train)

best_index = grid_search.best_index_                              #To take the standard deviation
best_std = grid_search.cv_results_['std_test_score'][best_index]  #To take the standard deviation of the Cross-Validation (5-Fold)

#To print the Result
print("-" * 27)
print(f"Best RF AUC = {grid_search.best_score_:.4f} (+/- {best_std * 2:.4f})")
print(f"Best parameters: {grid_search.best_params_}")

#To use the best model and i assign to rf
rf = grid_search.best_estimator_

```

**Implementation** of Random forest, testing more parameters, to optimize the parameters for Random Forest. **The standard deviation (std)** remains consistently low (**+/- 0.02**) across nearly all tests. This indicates that the model is **highly stable:** its performance does not vary drastically depending on how the data is split between training and validation.

#

#### AdaBoost

```python
#SEARCH FOR THE BEST PARAMETER FOR ADABOOST
"""AdaBoost parameter optimization. The code iterates through various combinations
of number of estimators (n_estimators) and learning rates (lerning_rates),evaluating each configuration with
Cross-Validation (ROC-AUC metric). Finally, the model is retrained with the best configuration found."""

print("----- ADABOOST OPTIMIZATION  -----")

#Define the base model and fix random_state for reproducibility
ada_base = AdaBoostClassifier(random_state=42)

#Define the parameters that i want to try
param_grid = {
    'learning_rate': [0.01, 0.1, 1.0],
    'n_estimators': [50, 100, 200]
}

#Implementation of GridSearch
gs = GridSearchCV(
    estimator=ada_base,
    param_grid=param_grid,
    cv=5,                 #Evaluate with Cross-Validation (5-Fold)
    scoring='roc_auc',    #I use ROC as metrics of valutation
    n_jobs=-1,            #To velocize the process
    refit=True            #Retrain the best at the end
)

#To fit the best model
gs.fit(X_train, y_train)

best_index = grid_search.best_index_                              #To take the standard deviation
best_std = grid_search.cv_results_['std_test_score'][best_index]  #To take the standard deviation of the Cross-Validation (5-Fold)

#To print the Result
print(f"Best AB AUC = {grid_search.best_score_:.4f} (+/- {best_std * 2:.4f})")
print(f"Best parameters: {grid_search.best_params_}")

#To use the best model and i assign to ada
ada = gs.best_estimator_

```

**Implementation** of AdaBoost, testing more parameters, to optimize the parameters for Adaboost. **The standard deviation (std)** remains consistently low (**+/- 0.02**) across nearly all tests. This indicates that the model is **highly stable:** its performance does not vary drastically depending on how the data is split between training and validation.

#

#### Extra Trees

```python
#SEARCH FOR THE BEST PARAMETER FOR EXTRA TREES
"""Extra Trees parameter optimization. The code iterates through various combinations
of tree counts (n_estimators), depth (max_depth) and split criteria (criterion), evaluating each configuration with
Cross-Validation (ROC-AUC metric). The class_weight='balanced' parameter is used to handle class imbalance. Finally, the model is retrained with
the best configuration found."""

print("----- EXTRA TREES OPTIMIZATION  -----")

#Define the base model and fix random_state for reproducibility
et_base = ExtraTreesClassifier(
    random_state=42,
    class_weight='balanced'   #i use Balanced to force the model to give more importance to the people that will leave
)

#Define the parameters that i want to try
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

#Implementation of GridSearch
grid_search = GridSearchCV(
    estimator=et_base,
    param_grid=param_grid,
    cv=5,                           #Evaluate with Cross-Validation (5-Fold)
    scoring='roc_auc',              #I use ROC as metrics of valutation
    n_jobs=-1,                      #To velocize the process
    verbose=1,                      #To give the information during the process
    refit=True                      #Retrain the best at the end
)

#To fit the best model
grid_search.fit(X_train, y_train)

best_index = grid_search.best_index_                              #To take the standard deviation
best_std = grid_search.cv_results_['std_test_score'][best_index]  #To take the standard deviation of the Cross-Validation (5-Fold)

#To print the Result
print("-" * 30)
print(f"Best ET AUC = {grid_search.best_score_:.4f} (+/- {best_std * 2:.4f})")
print(f"Best parameters: {grid_search.best_params_}")

#To use the best model and i assign to et
et = grid_search.best_estimator_
```

**Implementation** of Extra Trees, testing more parameters, to optimize the parameters for Extra Trees. **The standard deviation (std)** remains consistently low (**+/- 0.02**) across nearly all tests. This indicates that the model is **highly stable:** its performance does not vary drastically depending on how the data is split between training and validation.

#

#### Stacking

```python
#STACKING CLASSIFIER IMPLEMENTATION
"""Stacking Classifier implementation. This ensemble technique combines the predictions of the previously optimized 
base models (Random Forest, Gradient Boosting, AdaBoost, Decision Tree, Extra Trees) using a meta-learner (Logistic Regression).
The goal is to leverage the strengths of each individual model to improve overall performance."""

print("----- STACKING CLASSIFIER TRAINING -----")

#Define the base learners using the best estimators found via GridSearch
estimators = [
    ('rf', rf),
    ('gb', gb),
    ('ada', ada),
    ('et', et),
]

#Define the Stacking Classifier
#final_estimator is the meta-model that learns how to best combine the base models predictions
stacking_clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    cv=5,        #Evaluate with Cross-Validation (5-Fold)
    n_jobs=-1    ##To velocize the process
)

#To fit the best model
stacking_clf.fit(X_train, y_train)

best_index = grid_search.best_index_                              #To take the standard deviation
best_std = grid_search.cv_results_['std_test_score'][best_index]  #To take the standard deviation of the Cross-Validation (5-Fold)

#To print the Result
print("-" * 30)
print(f"Best Stacking AUC = {grid_search.best_score_:.4f} (+/- {best_std * 2:.4f})")
print(f"Best parameters: {grid_search.best_params_}")

#To use the best model and i assign to stack
stack = stacking_clf
```

**Implementation** of Stacking, testing more parameters, to optimize the parameters for Stacking. **The standard deviation (std)** remains consistently low (**+/- 0.02**) across nearly all tests. This indicates that the model is **highly stable:** its performance does not vary drastically depending on how the data is split between training and validation.

#### Prediction on Test Set

```python
#Predictions on Test Set 

y_prob_rf = rf.predict_proba(X_test)[:, 1]       #This predict the probability on the Test Set for Random Forest.
y_prob_gb = gb.predict_proba(X_test)[:, 1]       #This predict the probability on the Test Set for Gradient Boosting.  
y_prob_ada = ada.predict_proba(X_test)[:, 1]     #This predict the probability on the Test Set for AdaBoost 
y_prob_et = et.predict_proba(X_test)[:, 1]       #This predict the probability on the Test Set for Extra Trees
y_prob_stack = stack.predict_proba(X_test)[:, 1] #This predict the probability on the Test Set for Stacking


y_pred_rf = rf.predict(X_test)       #This one instead say only 0 or 1, of the Target for Random Forest.
y_pred_gb = gb.predict(X_test)       #This one instead say only 0 or 1, of the Target for Gradient Boosting.
y_pred_ada = ada.predict(X_test)     #This one instead say only 0 or 1, of the Target for AdaBoost
y_pred_et = et.predict(X_test)       #This one instead say only 0 or 1, of the Target for Extra Trees
y_pred_stack = stack.predict(X_test) #This one instead say only 0 or 1, of the Target for Stacking
```

Implementation of prediction on Test Set.

#

#### Computation of the metrics

```python
#COMPUTATION OF THE METRICS 
"""Comparative performance evaluation using Accuracy and AUC-ROC metrics. The code calculates false/true positive 
rates (FPR, TPR) for ROC curves and implements an automatic selection logic to declare the 'Best Model' based on the highest AUC score"""

acc_rf = accuracy_score(y_test, y_pred_rf)        #Accuracy of Random forest
acc_gb = accuracy_score(y_test, y_pred_gb)        #Accuracy of Gradient Boosting
acc_ada = accuracy_score(y_test, y_pred_ada)      #Accuracy of Adaboost
acc_et = accuracy_score(y_test, y_pred_et)        #Accuracy of Extra Trees
acc_stack = accuracy_score(y_test, y_pred_stack)  #Accuracy of Stacking


#I compute the necessary data for ROC Curve
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)          #To find False Positive Rate and True Positive Rate for Random Forest
fpr_gb, tpr_gb, _ = roc_curve(y_test, y_prob_gb)          #To find False Positive Rate and True Positive Rate for Gradient Boost
fpr_ada, tpr_ada, _ = roc_curve(y_test, y_prob_ada)       #To find False Positive Rate and True Positive Rate for AdaBoost
fpr_et, tpr_et, _ = roc_curve(y_test, y_prob_et)          #To find False Positive Rate and True Positive Rate for Extra Trees
fpr_stack, tpr_stack, _ = roc_curve(y_test, y_prob_stack) #To find False Positive Rate and True Positive Rate for Stacking

#Computation of AUC, area under curve
auc_rf = auc(fpr_rf, tpr_rf)           #Random forest
auc_gb = auc(fpr_gb, tpr_gb)           #Gradient Boosting
auc_ada = auc(fpr_ada, tpr_ada)        #AdaBoost
auc_et = auc(fpr_et, tpr_et)           #Extra Trees
auc_stack = auc(fpr_stack, tpr_stack)  #Stacking

#To print the metrics
print(f"Random Forest -> AUC: {auc_rf:.3f}")
print(f"Random Forest OOB Score: {rf.oob_score_:.2%} (Internal Validation)")
print(f"Gradient Boosting -> AUC: {auc_gb:.3f}")
print(f"AdaBoost -> AUC: {auc_ada:.3f}")
print(f"Extra Trees -> AUC: {auc_et:.3f}")
print(f"Stacking -> AUC: {auc_stack:.3f}")

#Find the best Model
models_auc = {
    'Random Forest': auc_rf,
    'Gradient Boosting': auc_gb,
    'AdaBoost': auc_ada,
    'Extra Trees': auc_et,
    'Stacking': auc_stack
}
best_model_name = max(models_auc, key=models_auc.get)   #To get the max of AUC from our model, i get the best model
best_model_auc = models_auc[best_model_name]            #To get after the value of AUC

#To print the advice
print("-" * 29)
print(f"The best model is {best_model_name} with an AUC of {best_model_auc:.3f}.")
```

**AUC** ensures the model effectively **separates churners from loyal customers**, preventing false conclusions caused by the unbalanced dataset.

#### ROC Curve Comparison

```python
# Plot ROC Curve Comparison
"""Comparative graphic visualization of ROC curves for Random Forest, Gradient Boosting and AdaBoost models.. 
The plot overlays the performance of both classifiers against the dashed diagonal line, 
representing the threshold of randomness (no predictive capability)."""



plt.figure(figsize=(8, 5))
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.3f})', linewidth=2)
plt.plot(fpr_gb, tpr_gb, label=f'Gradient Boosting (AUC = {auc_gb:.3f})', linewidth=2)
plt.plot(fpr_ada, tpr_ada, label=f'AdaBoost (AUC = {auc_ada:.3f})', linewidth=2)
plt.plot(fpr_et, tpr_et, label=f'Extra Trees (AUC = {auc_et:.3f})', linewidth=2)
plt.plot(fpr_stack, tpr_stack, label=f'Stacking (AUC = {auc_stack:.3f})', linewidth=2 )

plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

The **Random Forest** curve (blue) consistently outperforms compared to the majority, but **falls** slightly below the **Gradient Boosting** (orange), and **Stacking** (purple). The **marginal gain** in predictive power does not justify the increased complexity, making **Random Forest** the more **efficient** and interpretable choice.

#### Report

```python
#TO PLOT THE REPORT OF EVERY METHOD TO SEE THE GENERAL METRICS (PRECISION, RECALL, F1-SCORE)
"""Generation and tabular formatting of the Classification Report. Detailed metrics (Precision, Recall, F1-Score) 
are calculated for each class ('Loyal' and 'Churn'). The report is converted into a DataFrame and 'cleaned' by removing 
accuracy, macro avg,weighted avg. """


#I generate the report as dictionary to drop the useless rows
report_dict_rf = classification_report(y_test, y_pred_rf, target_names=['Loyal', 'Churn'], output_dict=True)
report_dict_gb = classification_report(y_test, y_pred_gb, target_names=['Loyal', 'Churn'], output_dict=True)
report_dict_stack = classification_report(y_test, y_pred_stack, target_names=['Loyal', 'Churn'], output_dict=True)
report_dict_et = classification_report(y_test, y_pred_et, target_names=['Loyal', 'Churn'], output_dict=True)
report_dict_ada = classification_report(y_test, y_pred_ada, target_names=['Loyal', 'Churn'], output_dict=True)


#Transform it into a pandas dataframe to manipulate it better
report_rf = pd.DataFrame(report_dict_rf).transpose()           #Random Forest
report_gb = pd.DataFrame(report_dict_gb).transpose()           #Gradient Boosting
report_stack = pd.DataFrame(report_dict_stack).transpose()     #Stacking
report_et = pd.DataFrame(report_dict_et).transpose()           #Extra Trees
report_ada = pd.DataFrame(report_dict_ada).transpose()         #AdaBoost


#I remove the lines I'm not interested in
report_clean_rf = report_rf.drop(['accuracy', 'macro avg', 'weighted avg'])
report_clean_gb = report_gb.drop(['accuracy', 'macro avg', 'weighted avg'])
report_clean_stack = report_stack.drop(['accuracy', 'macro avg', 'weighted avg'])
report_clean_et = report_et.drop(['accuracy', 'macro avg', 'weighted avg'])
report_clean_ada = report_ada.drop(['accuracy', 'macro avg', 'weighted avg'])


#To print the result of the report
print("----- Report for Random Forest -----")
print(report_clean_rf.round(3))

print("----- Report for Gradient Boosting -----")
print(report_clean_gb.round(3))

print("-------- Report for Stacking -----------")
print(report_clean_stack.round(3))

print("-------- Report for Extra Trees -----------")
print(report_clean_et.round(3))

print("-------- Report for AdaBoost -----------")
print(report_clean_ada.round(3))
```

The **Random Forest** model demonstrates excellent performance in identifying loyal customers. The **Recall of 63.6%** shows that the system captures the majority of actual churners, although it still misses a non-negligible portion. Overall, the model adopts a **"conservative"** approach: it ensures high confidence when making a positive prediction but tends to overlook the most uncertain cases.

## FEATURE IMPORTANCE

```python
#UNDERSTANDING THE CAUSES OF THE ABANDONMENT RATE (VARIABLES)
"""Feature Importance analysis for model interpretability. Importance coefficients are extracted from the Random Forest
to identify which variables most significantly influence the churn decision. Results 
are ranked in descending order, listing the 'Top 5 Risk Factors' and visualizing the impact of all variables via a bar plot."""


print("----- WHAT DRIVES CHURN? (Feature Importance) -----")

importances = rf.feature_importances_     #Is to assign the probability to every variable, but we can't figure out who it belongs to
feature_names = X_train.columns           #Retrieve the column names from your training dataset, we need it to give a name to those numbers from before
indices = np.argsort(importances)[::-1]   #Sort descending, from the more important to less important

#To print the importance of the first five Factors
print("First 5 Risk Factors:")
for i in range(5):
    print(f"{i+1}. {feature_names[indices[i]]}: {importances[indices[i]]:.4f}")

#To plot feature importance
plt.figure(figsize=(10, 5))
sns.barplot(x=importances[indices], y=feature_names[indices], hue=feature_names[indices], legend=False, palette="viridis")
plt.title("Which variables influence churn the most?")
plt.xlabel("Importance (Mean Decrease in Impurity)")
plt.show()
```

## FAIRNESS ANALYSIS

#### Independence

```python
#INDEPENDENCE COMPUTATION
"""Independence check. The code simply calculates the percentage of churn predictions 
for men and compares it with that for women. If the two percentages are similar, the model is fair; if they are very different, 
it might be favoring or penalizing one gender"""


#Create a temporary dataframe to facilitate calculations
res = X_test.copy()                 #To create an exact copy of Test Set
res['Truth'] = y_test               #Add a new column called "Truth" containing the real values
res['Prediction'] = y_pred_rf       #Add the column with the model decisions
res['Probability'] = y_prob_rf      #Add the column with the risk percentage
res['Gender'] = S_test              #Add the Gender column

#Independence computation
#Does the model predict the same churn rate for men and women?
pos_rate_m = res[res['Gender'] == 'Male']['Prediction'].mean() #Take my results table (res) and filter keeping only the rows where the customer is "Male" and take the mean
pos_rate_f = res[res['Gender'] == 'Female']['Prediction'].mean() #Same thing as above, but for Female.
print(f"[Independence] Predicted Churn Rate:")
print(f"Male   : {pos_rate_m:.1%}")
print(f"Female : {pos_rate_f:.1%}")
```

Analyzing the **Independence metric**, there is a clear **disparity** in the predicted churn rates between the two demographic groups: the rate for females (**26.7%**) is significantly higher than that for males (**16.7%**). This substantial difference indicates that the model **violates the Independence criterion**, suggesting a potential bias where female customers are classified as 'at risk' of churning more frequently than male customers.

#

#### Separation

```python
#SEPARATION 
#  If a customer actually leaves, do we identify them with equal precision for both genders?
"""Verification of Separation.We calculate if the model is equally good at 
identifying leavers (True Positive Rate) and how often it gives false alarms (False Positive Rate) in the same way for both 
men and women. If the numbers differ, the model is treating errors discriminatorily."""



def get_metrics(df_sub): #accept as un input a sub dataframe
    #Extract True Negatives, False Positives, False Negatives, True Positives
    tn, fp, fn, tp = confusion_matrix(df_sub['Truth'], df_sub['Prediction'], labels=[0,1]).ravel() #Use ravel to flatten the matrix(2x2) into a row of 4 numbers
    
    #Sensitivity = (True positive / True positive + false negative)
    tpr = tp / (tp + fn) if (tp + fn) > 0 else 0 
    
    #False positive Rate = ( false positive/ false positive + true negative )
    fpr = fp / (fp + tn) if (fp + tn) > 0 else 0 
    
    return tpr, fpr

# Calculate metrics for each group
tpr_m, fpr_m = get_metrics(res[res['Gender'] == 'Male'])
tpr_f, fpr_f = get_metrics(res[res['Gender'] == 'Female'])

print(f"[Separation] Error Metrics:")
print(f"Male   -> True Positive Rate: {tpr_m:.1%}, False Positive Rate: {fpr_m:.1%}")
print(f"Female -> True Positive Rate: {tpr_f:.1%}, False Positive Rate: {fpr_f:.1%}")
```

The **Separation criterion is not satisfied**: although the True Positive Rate is similar across groups, there is a **significant** disparity in the False Positive Rate (**13.4%** for females vs. **7.9%** for males), indicating that the model generates nearly twice as many false alarms for female customers.

#### Sufficiency

```python
#SUFFICIENCY 
#  "Does a 60% risk score mean 60% real risk for everyone?"

"""Verification of Sufficiency. Here we check if the risk percentages given by the model are 
true for everyone. If the model says a group of people has a 60% probability of leaving, 60% of them should actually leave. 
The plot shows if this reliability holds true for both men and women"""



#Take as input Truth and the probability, only from male, and divide all the in five group (0-20%, 20%-40%....) and in every bucket
#compute the mean of predicted probabilities and count the percentage of the real probability, the truth
prob_true_m, prob_pred_m = calibration_curve(
    res[res['Gender'] == 'Male']['Truth'],
    res[res['Gender'] == 'Male']['Probability'],
    n_bins=5
)


#Same as above
prob_true_f, prob_pred_f = calibration_curve(
    res[res['Gender'] == 'Female']['Truth'],
    res[res['Gender'] == 'Female']['Probability'],
    n_bins=5
)

print("Remember: It is mathematically impossible to satisfy all 3 criteria simultaneously (Impossibility Theorem).")

# Plot Calibration Curve
plt.figure(figsize=(8, 5))
plt.plot(prob_pred_m, prob_true_m, marker='o', label='Male')
plt.plot(prob_pred_f, prob_true_f, marker='s', label='Female')
plt.plot([0, 1], [0, 1], 'k--', label='Perfect Calibration')
plt.xlabel('Mean Predicted Probability')
plt.ylabel('Actual Churn Rate')
plt.title('Sufficiency: Calibration Curve by Gender')
plt.legend()
plt.show()
```

Visual analysis demonstrates that the **Sufficiency criterion is fully satisfied.** The calibration curves for both groups (males and females) are almost perfectly superimposed and follow very closely the dashed diagonal representing ideal calibration ('Perfect Calibration').
